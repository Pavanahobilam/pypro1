TASK-1: Guessing Game

Write a Python program that generates a
random number between 1 and 100. The
user should then try to guess the number.
The program should provide hints such as
"too high" or "too low" until the correct
number is guessed.

```python
import random

def number_guessing_game(min_num, max_num):
    # Generate a random number between min_num and max_num
    secret_number = random.randint(min_num, max_num)

    # Initialize the number of attempts
    attempts = 0

    # Start the game loop
    while True:
        # Get the user's guess
        guess = int(input(f"Guess the number between {min_num} and {max_num}: "))

        # Increment the number of attempts
        attempts += 1

        # Check if the guess is correct
        if guess == secret_number:
            print(f"Congratulations! You guessed the number {secret_number} correctly in {attempts} attempts.")
            break
        elif guess < secret_number:
            print("Too low! please try again.")
        else:
            print("Too high!please try again.")

# Set the range for the random number
min_num = 1
max_num = 100

# Start the number guessing game
number_guessing_game(min_num, max_num)
```

Task-3: Password Strength Checker

Create a Python function that evaluates
the strength of a password entered by the
user. Implement checks for factors such as
length, presence of uppercase and
lowercase letters, digits, and special
characters.

```python
import string

def evaluate_password_strength(password):
    # Define strength criteria
    length_criteria = 8
    has_uppercase = any(char.isupper() for char in password)
    has_lowercase = any(char.islower() for char in password)
    has_digit = any(char.isdigit() for char in password)
    has_special = any(char in string.punctuation for char in password)

    # Evaluate strength based on criteria
    strength = 0
    if len(password) >= length_criteria:
        strength += 1
    if has_uppercase:
        strength += 1
    if has_lowercase:
        strength += 1
    if has_digit:
        strength += 1
    if has_special:
        strength += 1

    return strength

# Get password input from the user
password = input("Enter your password: ")

# Evaluate password strength
strength = evaluate_password_strength(password)

# Print strength evaluation
print("Password Strength:", strength)
```

Task-4: Fibonacci Sequence

Write a Python function that generates the
Fibonacci sequence up to a given number of
terms. The function should take an integer
input from the user and display the
Fibonacci sequence up to that number of
terms.

```python
def fibonacci_sequence(terms):
    list_elements = []
    # First two terms of the Fibonacci sequence
    a, b = 0, 1

    # Generate Fibonacci sequence up to the specified number of terms
    for _ in range(terms):
        list_elements.append(a)
        a, b = b, a + b

    return list_elements

# Get the number of terms from the user
num_terms = int(input("Enter the number of terms for the Fibonacci sequence: "))

# Generate and display the Fibonacci sequence
fib_sequence = fibonacci_sequence(num_terms)
print("Fibonacci Sequence:")
print(fib_sequence)
```

Task-5: File Manipulation

Write a Python program that reads a text
file and counts the occurrences of each
word in the file. Display the results in
alphabetical order along with their
respective counts.

```python
# importing a url file from https://www.gutenberg.org/cache/epub/73477/pg73477.txt
import requests
results= requests.get('https://www.gutenberg.org/cache/epub/73477/pg73477.txt')
text=results.text
```

```python
print(text)
```

```python
word_list=text.split(" ")
len(word_list)
```

```python
def count_word_occurrences(text):
    word_count = {}

    # Split the text into words
    words = text.split()

    # Iterate through each word
    for word in words:
        # Remove punctuation and convert to lowercase
        word = word.strip().lower().strip('.,!?')
        # Update the word count dictionary
        if word:
            word_count[word] = word_count.get(word, 0) + 1

    # Display the word count results in alphabetical order
    sorted_word_count = sorted(word_count.items(), key=lambda x: x[0])
    for word, count in sorted_word_count:
        print(f"{word}: {count}")



# Count word occurrences and display the results
count_word_occurrences(text)
```

Task-2: Temperature Conversion

Create a Python program that converts
temperatures between Celsius and
Fahrenheit. Prompt the user to enter a
temperature value and the unit of
measurement, and then display the
converted temperature.

```python
def celsius_to_fahrenheit(a):
    return (a * 9/5) + 32

def fahrenheit_to_celsius(b):
    return (b - 32) * 5/9

# Prompt the user to enter temperature and unit
temperature = float(input("Enter the temperature value: "))
unit = input("Enter the unit of measurement (C or F): ")

# Perform conversion based on the unit provided
if unit.upper() == 'C':
    converted_temperature = celsius_to_fahrenheit(temperature)
    print(f"{temperature}°C is equal to {converted_temperature}°F")
elif unit.upper() == 'F':
    converted_temperature = fahrenheit_to_celsius(temperature)
    print(f"{temperature}°F is equal to {converted_temperature}°C")
else:
    print("Invalid unit of measurement. Please enter 'C' for Celsius or 'F' for Fahrenheit.")
```

Task-3: Email Validator

Develop a Python function that validates
whether a given string is a valid email
address. Implement checks for the format,
including the presence of an "@" symbol and
a domain name

```python
import re

def validate_gmail(gmail):
    # Regular expression pattern for email validation
    pattern = r'^[a-zA-Z0-9._%+-]+@gmail.com'

    # Check if the email matches the pattern
    if re.match(pattern, gmail):
        return True
    else:
        return False

# Get input email address from the user
gmail_address = input("Enter your gmail address: ")

# Validate the email address
if validate_gmail(gmail_address):
    print("Valid gmail address.")
else:
    print("Invalid gmail address.")
```

Task-4: Calculator Program

Create a Python program that acts as a basic
calculator. It should prompt the user to
enter two numbers and an operator (+
, -,
*
, /,
%), and then display the result of the
operation.

```python
def add(a, b):
    return a + b

def sub(a, b):
    return a - b

def mul(a, b):
    return a * b

def div(a, b):
    return a / b

print("Select operation you want:")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")

while True:
    choice = input("Enter choice (1/2/3/4): ")

    if choice in ('1', '2', '3', '4'):
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        if choice == '1':
            print(num1, "+", num2, "=", add(num1, num2))
        elif choice == '2':
            print(num1, "-", num2, "=", sub(num1, num2))
        elif choice == '3':
            print(num1, "*", num2, "=", mul(num1, num2))
        elif choice == '4':
            if num2 == 0:
                print("Error! Division by zero is not allowed.")
            else:
                print(num1, "/", num2, "=", div(num1, num2))
    else:
        print("Invalid input")

    next_calculation = input("Do you want to do another calculation? (yes/no): ")
    if next_calculation.lower() != "yes":
        break
```

Task-5: Palindrome Checker

Write a Python function that checks whether
a given string is a palindrome. A palindrome
is a word, phrase, or sequence that reads the
same backward as forward (e.g.,

"madam" or

"racecar").

```python
def is_palindrome(s):
    # Remove spaces and convert to lowercase
    s = s.replace(" ", "").lower()

    # Check if the string is equal to its reverse
    return s == s[::-1]

# Test the function
string1 = input("Enter words separated by spaces: ")
print(f"'{string1}' is a palindrome:", is_palindrome(string1))

```
